```python
from sklearn.model_selection import train_test_split
```

```python
y=df["diagnosis"]
```

```python
y
```

```python
x=df.drop(columns=["radius_se",                  
"texture_se",                 
"perimeter_se",               
"area_se",                    
"smoothness_se",              
"compactness_se",             
"concavity_se",              
"concave points_se",          
"symmetry_se",                
"fractal_dimension_se","id"])
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.2)
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf= RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
y_test
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True,xticklabels=['Benign', 'Malignant'], yticklabels=['Benign', 'Malignant'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
y_pred1=rf.predict(x_train)
accuracy1 = accuracy_score(y_train, y_pred1)
report1 = classification_report(y_train, y_pred1)
#accuracy1
print(report1)
```
