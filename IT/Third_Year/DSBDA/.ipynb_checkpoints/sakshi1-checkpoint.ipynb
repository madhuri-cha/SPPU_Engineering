```python
import pandas as pd
```

```python
df=pd.read_csv("cancer_data_set.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df[['diagnosis']]
```

```python
df['diagnosis'] = df['diagnosis'].map({'M':1 , 'B':0})
```

```python
df['diagnosis']
```

```python
x = df.iloc [: , 2:]
```

```python
x
```

```python
y = df['diagnosis']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,X_test, y_train , Y_test = train_test_split(x , y , random_state = 42 , test_size = 0.2)
```

```python
x_train
```

```python
X_test
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf
```

```python
x_train.dtypes
```

```python
y_train.dtypes
```

```python
rf.fit(x_train , y_train)
```

```python
y_predict = rf.predict(x_train)
```

```python
y_predict
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_train,y_predict)
```

```python
from sklearn.metrics import classification_report
```

```python
from sklearn.metrics import confusion_matrix
```

```python
report = classification_report(y_train,y_predict)
```

```python
report
```

```python
matrix = confusion_matrix(y_train,y_predict)
```

```python
matrix
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
cm = ConfusionMatrixDisplay.from_predictions(y_train ,y_predict)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Select the first tree from the Random Forest
tree_to_plot = rf.estimators_[0]

# Plot the tree
plt.figure(figsize=(20, 10))
plot_tree(
    tree_to_plot,
    feature_names=x_train.columns,
    filled=True,
    rounded=True,
    class_names=['Benign', 'Malignant'],  # Replace with your class labels if needed
    max_depth=3  # Limit depth for readability (optional)
)
plt.title("Decision Tree from Random Forest")
plt.show()
```
